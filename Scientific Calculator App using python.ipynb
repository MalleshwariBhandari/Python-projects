```python
#importing modules
import random
import math
```

```python
# Function for finding percentage
def percentage(num,value):
        percent_output= float(num) *(value/100)
        print(value," % of ",num," = ",percent_output)
```

```python
# Function for finding factorial
def factorial(num):
        factorial = 1
        if num < 0:
           print("Sorry,factorial does not exist for negative numbers")
        elif num == 0:
           print("The factorial of 0 is 1")
        else:
           for i in range(1,num + 1):
               factorial = factorial*i
           print(num,"!","=",factorial)
```

```python
# Function for finding LCM
def LCM(num1,num2):
        if num1 > num2:  
           greater = num1  
        else:  
           greater = num2 
        while(True):  
           if((greater % num1 == 0) and (greater % num2 == 0)):  
                lcm_output = greater  
                break  
           greater =  greater + 1
        print("LCM of",num1,"and",num2,"=",lcm_output)
```

```python
# Function for finding HCF
def HCF(num1,num2):
        if(num1>num2):
            smaller = num2
        else:
            smaller = num1
        i = 1
        while(i<=smaller):
            if(num1%i==0 and num2%i ==0):
                hcf_output = i
            i= i+1
        print("HCF of",num1,"and",num2,"=",hcf_output)
```

```python
# Function for finding square root
def square_root(num):
        sqrt_output=math.sqrt(num)
        print("The square root of",num,"=",sqrt_output)
```

```python
# Function for finding log
def log(num):
        log_output=math.log10(num)        #log value for base 10
        print("The log of ",num," = ",log_output)
```

```python
# Function for finding exponential
def exp(num): # exp(x) value
        exp_output=math.exp(num)
        print("The exp(%d)" %num, "= ", exp_output)
```

```python
# Function for finding exponential
def power(num, power_val): 
        power_output=math.pow(num,power_val)
        print(power_val,"raised to the power of", num,"=", power_output)
```

```python
# Function for finding cosine
def sin(num): 
        sin_val=math.sin(num)
        print("The sin of",num,"=",sin_val)
```

```python
# Function for finding cos
def cos(num): 
        cos_val=math.cos(num)
        print("The cos of",num,"=",cos_val)
```
